```python
# Task - A
n, m = map(int, input().split())
parent = [i for i in range(n + 1)]
size = [1] * (n + 1)

def find(x):
    if parent[x] != x:
        parent[x] = find(parent[x])
    return parent[x]

def union(x, y):
    rootx = find(x)
    rooty = find(y)

    if rootx == rooty:
        return size[rootx]

    if size[rootx] < size[rooty]:
        rootx, rooty = rooty, rootx

    parent[rooty] = rootx
    size[rootx] += size[rooty]
    return size[rootx]

for i in range(m):
    u, v = map(int, input().split())
    print(union(u, v))
```

```python
import heapq
from collections import deque

n, m = map(int, input().split())
edges = []
adj = [[] for _ in range(n + 1)]

for _ in range(m):
    u, v, w = map(int, input().split())
    edges.append((w, u, v))
    adj[u].append((v, w))
    adj[v].append((u, w))

visited = [0] * (n + 1)
key = [float('inf')] * (n + 1)
parent = [-1] * (n + 1)
key[1] = 0

pq = [(0, 1)]
mst_edges = []
mst_cost = 0

while pq:
    w_u, u = heapq.heappop(pq)
    if visited[u]:
        continue
    visited[u] = 1
    mst_cost += w_u

    if parent[u] != -1:
        mst_edges.append((w_u, u, parent[u]))

    for v, w in adj[u]:
        if not visited[v] and w < key[v]:
            key[v] = w
            parent[v] = u
            heapq.heappush(pq, (w, v))

if len(mst_edges) != n - 1:
    print(-1)
    exit()

mst_adj = [[] for _ in range(n + 1)]
mst_set = set()

for w, u, v in mst_edges:
    mst_adj[u].append((v, w))
    mst_adj[v].append((u, w))
    mst_set.add((min(u, v), max(u, v), w))

def get_max_edge_in_path(u, v):
    visited_dfs = [False] * (n + 1)
    stack = [(u, -1)]
    parent = [-1] * (n + 1)
    max_edge = [0] * (n + 1)

    while stack:
        curr, curr_max = stack.pop()
        if visited_dfs[curr]:
            continue
        visited_dfs[curr] = True
        max_edge[curr] = curr_max

        if curr == v:
            return max_edge[v]

        for nei, w in mst_adj[curr]:
            if not visited_dfs[nei]:
                stack.append((nei, max(curr_max, w)))
                parent[nei] = curr

    return -1

second_mst_costs = []

for w, u, v in edges:
    edge_id = (min(u, v), max(u, v), w)
    if edge_id in mst_set:
        continue
    max_in_path = get_max_edge_in_path(u, v)
    if max_in_path == -1:
        continue
    new_cost = mst_cost - max_in_path + w
    if new_cost > mst_cost:
        second_mst_costs.append(new_cost)

if second_mst_costs:
    print(min(second_mst_costs))
else:
    print(-1)
```
